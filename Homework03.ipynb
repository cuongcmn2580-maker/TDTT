Bài 4: Viết chương trình in ra hình tam giác cân có độ cao h được nhập từ bàn phím 

```python
h = int(input())

for i in range(1, h + 1):
    for j in range(1, 2 * h):
        if i == h or i + j == h + 1 or j - i == h - 1:
            print("*", end="")
        else:
            print(" ", end="")
    print()
```

Bài 9: Nhập vào số nguyên dương a, kiểm tra xem a có phải là số chính phương hay
không? Số chính phương là số có căn bậc hai là một số nguyên VD: 1, 4, 9, 16, 25…. 


```python
a = int(input())

la_so_chinh_phuong = False
for i in range(1, a + 1):
    if i * i == a:
        la_so_chinh_phuong = True
        break

if la_so_chinh_phuong:
    print(a, "là số chính phương")
else:
    print(a, "không phải là số chính phương")
```

Bài 10: Nhập vào số nguyên dương a, kiểm tra xem a có phải là số hoàn hảo hay không.

Bài 12: Nhập số nguyên dương n, Cho S(k) = 1 + 2 + 3 + … + k. Tìm k sao cho S(k)
lớn nhất nhưng nhỏ hơn n 

```python
n = int(input("Nhập vào số nguyên dương n: "))

s = 0
k = 0

while s + (k + 1) < n:
    k = k + 1
    s = s + k

print(k)
```

Bài 13: Nhập vào số nguyên dương A, tìm n nhỏ nhất sao cho
1 + 1/2 + 1/3 + 1/4 + ... + 1/n > A

```python
a = int(input())

s = 0
n = 0

while s <= a:
    n = n + 1
    s = s + 1/n

print(n)
```

Bài 14: Nhập vào một dãy số nguyên, dừng nhập khi người dùng nhập -1. Sau khi nhập
xong, in số lớn nhất, số nhỏ nhất trong những số vừa nhập. KHÔNG ĐƯỢC dùng mảng 

```python
print("Nhập vào dãy số (Nhập -1 để dừng):")

so_dau_tien = int(input())

if so_dau_tien == -1:
    print("Bạn chưa nhập số nào ngoài số dừng.")
else:
    lon_nhat = so_dau_tien
    nho_nhat = so_dau_tien
    
    while True:
        n = int(input())
        if n == -1:
            break
        
        if n > lon_nhat:
            lon_nhat = n
        if n < nho_nhat:
            nho_nhat = n
            
    print("Số lớn nhất là:", lon_nhat)
    print("Số nhỏ nhất là:", nho_nhat)
```

Bài 15: Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số. Làm bằng nhiều
cách nhất có thể, ko dùng mảng. 


```python
n = int(input("Nhập n: "))
tam = abs(n) 
dem = 0

if tam == 0:
    dem = 1
else:
    while tam > 0:
        tam = tam // 10
        dem = dem + 1

print("Số chữ số là:", dem)
```

Bài 16: Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số chẵn, bao nhiêu
chữ số lẻ 

```python
n = int(input("Nhập vào số nguyên dương n: "))

chan = 0
le = 0
tam = abs(n)

if tam == 0:
    chan = 1
else:
    while tam > 0:
        chu_so = tam % 10
        if chu_so % 2 == 0:
            chan = chan + 1
        else:
            le = le + 1
        tam = tam // 10

print("Số chữ số chẵn là:", chan)
print("Số chữ số lẻ là:", le)
```

Bài 17: Nhập vào số nguyên dương n, tính tổng các chữ số của n 

```python
n = int(input("Nhập vào số nguyên dương n: "))

tong = 0
tam = abs(n)

while tam > 0:
    tong = tong + (tam % 10)
    tam = tam // 10

print("Tổng các chữ số của n là:", tong)
```

Bài 18: Nhập vào một số nguyên dương n, kiểm tra xem n có phải là số dạng 3^k hay
không 


```python
n = int(input("Nhập vào số nguyên dương n: "))

is_3k = False
if n > 0:
    tam = n
    while tam % 3 == 0:
        tam = tam // 3
    
    if tam == 1:
        is_3k = True

if is_3k:
    print(n, "có dạng 3^k")
else:
    print(n, "không phải dạng 3^k")
```

Bài 19: Dãy số fibonacci là dãy số được định nghĩa như sau: 1, 1, 2, 3, 5, 8, 13,... với số
kế tiếp sẽ bằng tổng hai số trước đó. Nhập vào số nguyên dương A, hãy tìm số trong
dãy số fibonacci lớn nhất nhưng không vượt quá A 


```python
a = int(input("Nhập vào số nguyên dương A: "))

f1 = 1
f2 = 1
fib_tiep_theo = f1 + f2

if a < 1:
    print("Không có số Fibonacci nào.")
elif a == 1:
    print("Số Fibonacci lớn nhất không vượt quá A là: 1")
else:
    ket_qua = 1
    while fib_tiep_theo <= a:
        ket_qua = fib_tiep_theo
        f1 = f2
        f2 = fib_tiep_theo
        fib_tiep_theo = f1 + f2
    print("Số Fibonacci lớn nhất không vượt quá A là:", ket_qua)
```

Bài 20: Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số là số nguyên tố.

```python
n = int(input("Nhập vào số nguyên dương n: "))

dem = 0
tam = abs(n)

while tam > 0:
    chu_so = tam % 10
    # Các số nguyên tố có 1 chữ số là 2, 3, 5, 7
    if chu_so == 2 or chu_so == 3 or chu_so == 5 or chu_so == 7:
        dem = dem + 1
    tam = tam // 10

print("Số lượng chữ số là số nguyên tố là:", dem)
```
